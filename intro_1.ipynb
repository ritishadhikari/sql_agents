### Join Track and Media Type

```python
sql="""
    SELECT
        t.name AS track_name,
        t.composer,
        milliseconds AS duration_in_ms,
        t.unitprice,
        m.name AS media_type
    FROM
        track AS t
    INNER JOIN
        mediatype AS m
    WHERE 
        track_name='For Those About To Rock (We Salute You)'
"""
mediatypeTrack=sq(string=sql)
mediatypeTrack.rename(columns=lambda x:x.replace("_"," ").upper())
```

```python
sql=""" 
WITH cte AS (
    SELECT
        avg(total) AS averageTotal
    FROM
        invoice
)
SELECT
    I.*
FROM
    invoice AS I
    JOIN
    cte
WHERE
    I.total<cte.averageTotal
"""

sq(string=sql)
```

```python
sql="""
    SELECT
        billingCountry,
        CASE
            WHEN total>0 AND total<=7 THEN '0-10'
            WHEN total>7 AND total<=15 THEN '10-15'
            WHEN total>15 AND total<=26 THEN '15-26'
            ELSE 'NONE'
        END AS
            total
    FROM
        invoice
"""
sq(string=sql).head(10)
```

```python
sql=""" 
    SELECT
        billingcity,
        sum(total) AS sum_of_total
    FROM
        invoice
    GROUP BY 
        billingcity
    ORDER BY 
        sum(total)
    LIMIT
        10;
"""
sq(string=sql)
```
